# columns description

```python
df.columns
```

- **ID:** Unique record number.  
- **Distributor Name:** Name of the car distributor.  
- **Location:** Sale or distributor location.  
- **Car Name:** Model of the car.  
- **Manufacturer Name:** Brand or maker of the car.  
- **Car Type:** Category (e.g., SUV, Sedan).  
- **Color:** Car’s exterior color.  
- **Gearbox:** Transmission type (Manual/Automatic).  
- **Number of Seats:** Total seats.  
- **Number of Doors:** Total doors.  
- **Energy:** Fuel type (Petrol, Diesel, Electric, etc.).  
- **Manufactured Year:** Year the car was made.  
- **Price:** Listed or purchase price in USD.  
- **Mileage-KM:** Distance driven in kilometers.  
- **Engine Power-HP:** Engine power in horsepower.  
- **Purchased Date:** Date car was bought.  
- **Car Sale Status:** Availability (Sold/Available).  
- **Sold Date:** Date car was sold.  
- **Purchased Price:** Purchase cost in USD.  
- **Sold Price:** Sale price in USD.  
- **Margin-%:** Profit percentage.  
- **Sales Agent Name:** Salesperson’s name.  
- **Sales Rating:** Agent’s performance rating.  
- **Sales Commission-$:** Commission earned in USD.  
- **Feedback:** Customer comments.

# Adding Days to Sell column

```python
df["Days to sell"] = df["Sold Date"] - df["Purchased Date"]
df[["Purchased Date", "Sold Date", "Days to sell"]] 
```

```python
df.loc[df["Car Sale Status"]=="Un Sold","Days to sell"]=0
df
```

# Calculate Margin Amount && Final Price

```python
df["Margin Amount"] = df["Sold Price-$"] - df["Purchased Price-$"]
df 
```

```python
df.loc[df["Car Sale Status"]=="Un Sold","Margin Amount"]=0
df
```

```python
df["Final price"] = df["Sold Price-$"] - (df["Margin Amount"] + df["Sales Commission-$"])
df 
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

Sales_Agent = df.groupby("Sales Agent Name")

margin = Sales_Agent["Margin Amount"].sum().sort_values(ascending=False)
commission = Sales_Agent["Sales Commission-$"].sum().sort_values(ascending=False)

fig = make_subplots(
    rows=1,
    cols=2,
    specs=[[{"type": "xy"}, {"type": "xy"}]],
    subplot_titles=("Total Margin by Agent", "Total Commission by Agent")
)

# Margin Bar
fig.add_trace(
    go.Bar(
        x=margin.index,
        y=margin.values,
        name="Margin Amount",
        marker_color="#faedcd"
    ),
    row=1, col=1
)

# Commission Bar
fig.add_trace(
    go.Bar(
        x=commission.index,
        y=commission.values,
        name="Sales Commission-$",
        marker_color="#d4a373"
    ),
    row=1, col=2
)

fig.update_layout(
    template="simple_white",
    title="Sales Agent Performance Comparison",
    showlegend=True,
    xaxis=dict(tickangle=90),
    xaxis2=dict(tickangle=90),
    width=1000,   # equivalent to figsize width=10
    height=600    # equivalent to figsize height=6

)

fig.show()
```
